## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
results = engine.execute(text("select distinct launch_site from flights;"))
# for row in results:
#     print(f'{row}')

res_df = pd.DataFrame(results)
res_df
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
results = engine.execute(text("select launch_site from flights where launch_site like 'CCA%' limit 5;"))
# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(results)
res_df
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
results = engine.execute(text("select sum(payload_mass_kg) from flights where customer = 'NASA (CRS)';"))
for row in results:
    print(f'Total payload mass carried by boosters launched by Nasa (CRS): {row[0]}kg')
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
results = engine.execute(text("select avg(payload_mass_kg) from flights where booster_version = 'F9 v1.1';"))
for row in results:
    print(f'Average payload mass carried by boosters version F9 v1.1: {float(row[0])}kg')
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
results = engine.execute(text("""
    select landing_outcome,min(date) 
    from flights 
    where landing_outcome like '%ground pad%'
    group by landing_outcome;
    """))
# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(results)
res_df
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
results = engine.execute(text("""
    select booster_version, payload_mass_kg, landing_outcome
    from flights
    where landing_outcome like '%Success%%drone ship%'
    and payload_mass_kg between 4000 and 6000;
"""))
# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(results)
res_df
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
results = engine.execute(text("""
    select count(landing_outcome) 
    from flights
    where landing_outcome like '%Success%'
    or landing_outcome like '%Failure%';
    """))
for row in results:
    print(f'The total number of successful and failure mission outcomes: {row[0]}')
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
results = engine.execute(text("""
    select f1.booster_version , f1.payload_mass_kg
    from flights f1
    where f1.payload_mass_kg = (
        select max(f2.payload_mass_kg) 
        from flights f2
    );
"""))

# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(data=results)
res_df
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
results = engine.execute(text("""
    select date, launch_site, booster_version, landing_outcome
    from flights 
    where extract(year from date) = 2015
    and landing_outcome like '%Failure%';
    """))

# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(data=results)
res_df
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
results = engine.execute(text("""
    select 
    rank() over( order by landing_outcome desc) as rank, 
    landing_outcome,
    count(landing_outcome) as landings,
    count(date) as "2010_to_2017"
    from flights
    where landing_outcome like 'Success (ground pad)'
    or landing_outcome like 'Failure (drone ship)'
    and extract(year from date) between 2010 and 2017
    group by landing_outcome
    order by rank;
    """))

# for row in results:
#     print(f'{row}')
res_df = pd.DataFrame(data=results)
res_df
```
